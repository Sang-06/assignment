*Rule based*

Defination- A system that works usng predefined rules written by humans. "If this happens-then do that" the logic is programed manually.

Learning no ability, fully dependent on human rule.
Easy to understand, logic is simple no data training required. It cannot adapt or improve,it breaks with unexpected inputs.

Example- spam filter
If email contain 'win money' - mark as spam

*Machine Learning*

Defination - A system that learns patterns from data instead of being explicitly programmed. Feed data--model learn pattern--make prediction and it improves with more data.

Learns automatically from data,It adapts new situation and handles omplex probelms. More accurate for real world problems, it can handle large and messy data.Also, need a lot of data as it is less transparent.

Example- a. Netflix recommendations b. Face recognition.
